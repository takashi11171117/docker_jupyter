```python
import pymysql
from pandas.io import sql
import pandas as pd
import seaborn as sns
sns.set(font='Ricty Diminished',font_scale=2)
import matplotlib.pyplot as plt
%matplotlib inline
pd.set_option("display.max_colwidth", 200)
pd.set_option("display.max_rows", 101)
connection = pymysql.connect(host='db', db='kanpari', user='root', password='password',charset="utf8",cursorclass=pymysql.cursors.DictCursor)
```

```python
fishingpost_terms = ['fish', 'howto', 'area']
fishingpost_dfs = {'fish':None,'howto':None,'area':None}
```

```python
for i in range(len(fishingpost_terms)):
    sql_str = '''
    SELECT
    term_taxonomy_id, name
    FROM wp_term_taxonomy wtt
    INNER JOIN wp_terms wt
    ON wtt.term_id = wt.term_id
    WHERE wtt.taxonomy = 'tax'
    '''.format(tax=fishingpost_terms[i])
    df = sql.read_sql(sql_str, connection)
    for index, row in df.iterrows():
        sql_str = '''
        SELECT
        COUNT(*) 全期間投稿数
        FROM wp_posts po
        INNER JOIN wp_term_relationships wtr
        ON po.ID = wtr.object_id
        WHERE wtr.term_taxonomy_id = {fish}
        '''.format(fish=row[0])
        if index == 0:
            df2 = sql.read_sql(sql_str, connection)
            df2['魚種'] = [row[1]]
        else:
            df3 = sql.read_sql(sql_str, connection)
            df3['魚種'] = [row[1]] 
            df2 = pd.concat([df2, df3])
    df2 = df2.ix[:,['魚種','全期間投稿数']]
    fishingpost_dfs[fishingpost_terms[i]] = df2
```

```python
for i in range(len(fishingpost_terms)):
    fishingpost_dfs[fishingpost_terms[i]].plot(kind="bar", title='全期間魚種', figsize=(40, 8), x=df2.columns[0],  y=df2.columns[1])
```
